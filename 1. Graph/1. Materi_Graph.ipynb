# Introduction Graph Theory


**PENGANTAR GRAPH**

---



1. Graph merupakan kumpulan verteks yang dihubungkan satu sama lainnya melalui sisi atau busur atau yang bisa dikenal dengan (Edges).


> a. suatu graph (G) terdiri dari dua himpunan yaitu himpunan verteks dan himpunan edge.

> b. Vertek merupakan elemen dari Graph (G) yang dapat disajikan dengan titik , lingkaran kecil atau node. 

> c. Sedangkan Edge merupakan elemen dari Graph (G) yang disajikan dengan garis. Edge dapat menunjukkan hubungan (relasi) sembarang. Seperti : Rute penerbangan, jalan raya, sambungan telepon, available bandwidth, dll. Contoh Notasi : G (V, E)

2. Graph digunakan untuk merepresentasikan objek-objek diskrit dan hubungan antar objek tersebut. 
3. Secara definitif : 
Graph G = (V,E), yang dalam hal ini : 


> a. V = Himpunan tidak kosong dari simpul-simpul (Vertices) ==> {v1, v2, ...., vn}

> b. E = Himpunan sisi (Edges) yang menghubungkan sepasang simpul ==> {e1, e2, ... ,en}

```python
'''Definifikan Graph dengan Hubungan masing-masing titik
1. Titik A terhubung dengan : titik c
2. Titik B terhubungan dengan : titik c, e
3. Titik C terhubung dengan : titik a, b, d, e
4. Titik D terhubung dengan : titik c
5. titik E terhubungan dengan : titik c, b
6. titik F terhubung dengan : - 
'''
graph_dict = { "a" : {"c"},  
          "b" : {"c", "e"},
          "c" : {"a", "b", "d", "e"},
          "d" : {"c"},
          "e" : {"c", "b"},
          "f" : {}
        }
```

```python
'''
Generate Edges dengan menghubungkan masing-masing titik/vertek
'''
def generate_edges(graph_dict):
  edges = []
  for vertex in graph_dict:
    for neighbour in graph_dict[vertex]:
      if {neighbour, vertex} not in edges:
        edges.append({vertex, neighbour})
  return edges

'''
fungsi yang mengenerate semua vertek dalam graph
Note : 
1. fungsi set() digunakan untuk hanya menampilkan nilai yang berbeda / tidak ada nilai yang sama,
   jika terdapat nilai yang sama hanya ditampilkan sekali
'''

def all_vertek(graph_dict): 
  vertek = set(graph_dict.keys()) #fungsi menggunakan set() lihat Note
  return vertek

print("Vertek Pada Graph : ", all_vertek(graph_dict))
print("Edge Pada Graph : ", generate_edges(graph_dict))
```

```python
'''
Mencari Node yang terisolasi
'''
def find_isolated_nodes(graph_dict):
    """ returns a set of isolated nodes. """
    isolated = set()
    for node in graph_dict:
        if not graph_dict[node]:
            isolated.add(node)
    return isolated
print("Node yang terisolasi : ",find_isolated_nodes(graph_dict))
```

```python
# Definisikan Graph Terlebih Dahulu
'''Definifikan Graph dengan Hubungan masing-masing titik
1. Titik A terhubung dengan : titik c
2. Titik B terhubungan dengan : titik c, e
3. Titik C terhubung dengan : titik a, b, d, e
4. Titik D terhubung dengan : titik c
5. titik E terhubungan dengan : titik c, b
6. titik F terhubung dengan : - 
'''
graph_dict = { "a" : {"c"},  
          "b" : {"c", "e"},
          "c" : {"a", "b", "d", "e"},
          "d" : {"c"},
          "e" : {"c", "b"},
          "f" : {}
        }
```

```python
'''
fungsi untuk menemukan jalur/path terpendek di dalam graph
'''
def find_shortest_path(graph, start, end, path =[]):
        path = path + [start]
        if start == end: #ketika start vertex = end vertek maka akan menereturn start vertex
            return path
        shortest = None
        for node in graph[start]: 
            if node not in path: 
                newpath = find_shortest_path(graph, node, end, path)
                if newpath:
                    if not shortest or len(newpath) < len(shortest):
                        shortest = newpath
        return shortest
print("Jalur dari Vertex d ke Vertek e : ", find_shortest_path(graph_dict, "d", "e"))
```
