**6. Transpose Image**



---

Kita dapat mengubah orientasi dari sebuah image menggunakan fungsi transpose. fungsi ini menggunakan salah satu parameter yang sudah tersedia. Diantaranya : 


*   Image.FLIP_LEFT_RIGHT(this inverse the rows of Image)

*   Image.FLIP_TOP_BOTTOM(this inverse the columns of Image)


*   Image.ROTATE_90 (this rotates an image by angle of 90)

*   Image.ROTATE_180 (this rotates an image by angle of 180)
*   Image.ROTATE_270 (this rotates an image by angle of 270)


*   Image.TRANSPOSE(This convert the rows into columns)

output dari fungsi ini akan mengembalikan object image







```python
# Sebelum dilakukan transpost

print("Gamabar Sebelum Dilakukan Transpost")
plt.imshow(image)
```

```python
# Setelah DI transpose

print("gambar setelah di transpose")

transpose_image1=image.transpose(Image.FLIP_LEFT_RIGHT)
transpose_image2=image.transpose(Image.FLIP_TOP_BOTTOM)
transpose_image3=image.transpose(Image.ROTATE_90)
transpose_image4=image.transpose(Image.ROTATE_180)
transpose_image5=image.transpose(Image.ROTATE_270)
transpose_image6=image.transpose(Image.TRANSPOSE)
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(transpose_image1)
plt.title("FLIP_LEFT RIGHT")
plt.subplot(3,2,2)
plt.imshow(transpose_image2)
plt.title("FLIP_TOP_BOTTOM")
plt.subplot(3,2,3)
plt.imshow(transpose_image3)
plt.title("ROTATE_90")
plt.subplot(3,2,4)
plt.imshow(transpose_image4)
plt.title("ROTATE_180")
plt.subplot(3,2,5)
plt.imshow(transpose_image5)
plt.title("ROTATE_270")
plt.subplot(3,2,6)
plt.imshow(transpose_image6)
plt.title("TRANSPOSE")
```

**7. Risizing Image**



---


> fungsi ini dapat merbah sebuah gambar (panjang dan lebar). fungsi ini menerima 2 parameter yang direpresentasikan sebagai pasangan tuple yang merupakan ukuran dari gambar baru, dimana output terakhir yaitu berupa objek gambar



> dalam resizing gambar dikenal sebuah algoritma interpolasi untuk mengurangi atau meningkatkan ukuran dari sebuah gambar. 

Berikut adalah beberapa teknik interpolasi : 

|Nama Teknik | Pengertian | 
|---|---|
| Nearest | memilih satu piksel terdekat dari gambar input. dan mengabaikan semua piksel lainnya. teknik sederhana, yang memiliki hasil agak buruk| 
| BOX | setiap piksel dari gambar awal akan berkontribusi pada image baru dengan menggunakan bobot tertentu. untuk upscaling hampir setara dengan metode Nearest |
| BILINER | untuk melakukan resizing dilakukan perhitungan nilai piksel gambar baru menggunakan interpolasi linear pada semua piksel yang dapat memberikan kontribusi terhadap nilai keluaran (gambar baru) |
| HAMMING | menghasilkan gambar yang jauh lebih tajam , tidak memiliki dislokasi ditingkat lokal seperti pada interpolasi BOX|
| BICUBIC | salah satu metode interpolasi yang lebih baik, lebih boros dari segi waktu, namun memiliki kualitas terbaik|
| LANCZOS | salah satu metode interpolasi yang lebih baik, lebih boros dari segi waktu, namun memiliki kualitas terbaik| 


```python
# Sebelum dilakukan Resize

print("Gamabar Sebelum Dilakukan Resize")
plt.imshow(image)
```

```python
newsize = (300, 300)
plt.figure(figsize=(10,10))
resized_image1=image.resize(newsize,Image.BILINEAR)
resized_image2=image.resize(newsize,Image.NEAREST)
resized_image3=image.resize(newsize,Image.BOX)
resized_image4=image.resize(newsize,Image.HAMMING)
resized_image5=image.resize(newsize,Image.BICUBIC)
resized_image6=image.resize(newsize,Image.LANCZOS)
# plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(resized_image1)
plt.title("BILINEAR")
plt.subplot(3,2,2)
plt.imshow(resized_image2)
plt.title("NEAREST")
plt.subplot(3,2,3)
plt.imshow(resized_image3)
plt.title("BOX")
plt.subplot(3,2,4)
plt.imshow(resized_image4)
plt.title("HAMMING")
plt.subplot(3,2,5)
plt.imshow(resized_image5)
plt.title("BICUBIC")
plt.subplot(3,2,6)
plt.imshow(resized_image6)
plt.title("LANCZOS")
```

**15. Alpha Blending**


---
untuk mencapurkan dua image menjadi tumpang tindih alpha blending terdiri atas 0 dan 1

jika mendekati 0 akan lebih banyak gambar 1
sedangkan, jika mendekati 1 akan lebih banyak ke gambar 2
terdapat rules :

*   make sure the image have a same size
*   make sure the image have the alpha channel (*.png format)

rumus ==> image1*(1.0- alpha) + image2 *alpha


```python
image_a = "/content/drive/MyDrive/Colab Notebooks/Modul Kecerdasan Buatan Pak Galih /4. Image Fundamental & Preprocessing/kucing_lucu/19.jpg"
image_b = "/content/drive/MyDrive/Colab Notebooks/Modul Kecerdasan Buatan Pak Galih /4. Image Fundamental & Preprocessing/kucing_lucu/22.jpg"
```

```python
img_1 = Image.open(image_a)
img_2 = Image.open(image_b)
print("Sebelum dilakukan Alpha Blending")
plt.subplot(2,2,1)
plt.imshow(img_1)
plt.subplot(2,2,2)
plt.imshow(img_2)
```

```python
print("Setelah dilakukan Alpha Blending")
image1 = img_1.copy()
image2 = img_2.copy()

# making both images same size
image2 = image2.resize(image1.size)

# blending
image_blend = Image.blend(image1, image2, 0.5)
plt.imshow(image_blend)
```

**16. RGB Composition**


---

sebuah image ada yang bertipe RGB dalam hal ini image tersebut terdiri atas 3 warna utama yaitu Red, Green, Blue.

```python
img_rgb = Image.open("/content/drive/MyDrive/Colab Notebooks/Modul Kecerdasan Buatan Pak Galih /4. Image Fundamental & Preprocessing/kucing_lucu/2.jpg")
plt.imshow(img_rgb)
```

```python
img_rgb_array = np.array(img_rgb)
```

```python
# Splitting Nomor 3
import numpy as np

fig, ax = plt.subplots(nrows= 3, ncols = 1, figsize=(10,10))

for c, ax in zip(range(3), ax):
  # membuat matrix zero
  split_img = np.zeros(img_rgb_array.shape, dtype = "uint8")

  # split di masing-masing channel 
  split_img[:, :, c] = img_rgb_array[:, :, c]

  # menampilkan setiap channel
  ax.imshow(split_img)
```

```python
# Melihat Komposisi RGB di masing-masing Channel


plt.figure(figsize=(20,20))
plt.subplot(3,1,1)
plt.title("R Channel")
plt.ylabel('Height : {}'.format(img_rgb_array.shape[0]))
plt.xlabel('Width : {}'.format(img_rgb_array.shape[1]))
plt.imshow(img_rgb_array[:,:,0])

plt.subplot(3,1,2)
plt.title("G Channel")
plt.ylabel('Height : {}'.format(img_rgb_array.shape[0]))
plt.xlabel('Width : {}'.format(img_rgb_array.shape[1]))
plt.imshow(img_rgb_array[:,:,1])

plt.subplot(3,1,3)
plt.title("B Channel")
plt.ylabel('Height : {}'.format(img_rgb_array.shape[0]))
plt.xlabel('Width : {}'.format(img_rgb_array.shape[1]))
plt.imshow(img_rgb_array[:,:,2])
plt.show()
```
